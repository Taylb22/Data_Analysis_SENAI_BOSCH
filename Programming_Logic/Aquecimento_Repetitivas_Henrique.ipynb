```python
# Tarefa 1

nivel_oleo = 55.0

limite_inferior = 50.0
limite_superior = 75.0

while nivel_oleo > 0:
    if nivel_oleo < limite_inferior:
        print(f"ALERTA CRÍTICO: Nível de óleo baixo {nivel_oleo:.1f}! Máquina será parada.")
        break
    elif nivel_oleo > limite_superior:
        print(f"ATENÇÃO: Nível de óleo alto {nivel_oleo:.1f}. Monitorando.")
    else:
        print(f"Nível de óleo OK: {nivel_oleo:.1f}%.")
    
    nivel_oleo = nivel_oleo - 1.0
        
```

```python
# Tarefa 2

lote_componentes = [9.8, 9.5, 9.9, 8.5, 9.7, 9.0, 6.0, 8.0, 9.6]

tolerancia_minima = 8.0
defeito_grave = 7.5

componentes_aprovados = 0
for componente in lote_componentes:
    if componente < defeito_grave:
        print(f"Componente com defeito grave detectado: {componente:.1f}. Interrompendo inspeção do lote.")
        break
    elif componente < tolerancia_minima:
        print(f"Componente {componente:.1f} fora da tolerância ideal, mas aceitável.")
    else:
        print(f"Componente {componente:.1f} dentro da tolerância. Contando como aprovado")
        componentes_aprovados += 1
else:
    print(f"total de componentes aprovados: {componentes_aprovados} de {len(lote_componentes)}")


print("Inspeção do lote finalizada.")
```
